```python
def add_filename(frame_no) :
    n_zeros = 6 - len(frame_no)
    f_name = 'frame_'
    for n in range(n_zeros) :
        f_name = f_name + '0'
    f_name = f_name + frame_no + '.jpg'
    return f_name

def get_xc(row,w=640,h=480) :
    dw = 1.0/w
    dh = 1.0/h
    xc = abs(row.x2+row.x1) / 2.
    yc = abs(row.y2+row.y1) / 2.
    w = abs(row.x2-row.x1)
    h = abs(row.y2-row.y1)
    xc = xc*dw
    w = w*dw
    yc = yc*dh
    h = h*dh
#     return xc,yc,w,h
    return xc

def get_yc(row,w=640,h=480) :
    dw = 1.0/w
    dh = 1.0/h
    xc = abs(row.x2+row.x1) / 2.
    yc = abs(row.y2+row.y1) / 2.
    w = abs(row.x2-row.x1)
    h = abs(row.y2-row.y1)
    xc = xc*dw
    w = w*dw
    yc = yc*dh
    h = h*dh
#     return xc,yc,w,h
    return yc

def get_w(row,w=640,h=480) :
    dw = 1.0/w
    dh = 1.0/h
    xc = abs(row.x2+row.x1) / 2.
    yc = abs(row.y2+row.y1) / 2.
    w = abs(row.x2-row.x1)
    h = abs(row.y2-row.y1)
    xc = xc*dw
    w = w*dw
    yc = yc*dh
    h = h*dh
#     return xc,yc,w,h
    return w

def get_h(row,w=640,h=480) :
    dw = 1.0/w
    dh = 1.0/h
    xc = abs(row.x2+row.x1) / 2.
    yc = abs(row.y2+row.y1) / 2.
    w = abs(row.x2-row.x1)
    h = abs(row.y2-row.y1)
    xc = xc*dw
    w = w*dw
    yc = yc*dh
    h = h*dh
#     return xc,yc,w,h
    return h
```

```python
traffic_df['filename'] = traffic_df['frame_no'].apply(add_filename)
# traffic_df[['xc', 'yc', 'w', 'h']] = traffic_df.apply(get_box_coordinates,axis=1,result_type = 'expand')
traffic_df['xc'] = traffic_df.apply(get_xc,axis=1)
traffic_df['yc'] = traffic_df.apply(get_yc,axis=1)
traffic_df['w'] = traffic_df.apply(get_w,axis=1)
traffic_df['h'] = traffic_df.apply(get_h,axis=1)
traffic_df
```
