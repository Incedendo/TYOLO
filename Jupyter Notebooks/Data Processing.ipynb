```python
import json

filepath = '/content/drive/MyDrive/Classroom/CSCE_452_643_Robotics_and_Spatial_Intelligence/Traffic_Light_Detection/Data Processing/instances_val2017.json'
# Opening JSON file
f = open(filepath,)
 
# returns JSON object as
# a dictionary
data = json.load(f)

w = 640
h  = 480

data.keys()

images = data['images']
print("no of images:", len(images))

annotations = data['annotations']
print("no of annotations:", len(annotations))
```

```python
img_dict = dict()
# Iterate over a sequence of numbers from 0 to 4
for img in images:
    img_id = img['id']
    img_dict[img_id] = img['file_name']

print("len of img_dict", len(img_dict.keys()))
```

```python
img_dict
```

```python
traffic_annotations = []

for ann in annotations:
    if ann["category_id"] == 10:
        traffic_annotations.append(ann)

print("number of traffic lights annotations", len(traffic_annotations))
```

```python
import pandas as pd
traffic_dict = dict()

# df = pd.dataframe()
with open('/content/drive/MyDrive/Classroom/CSCE_452_643_Robotics_and_Spatial_Intelligence/Traffic_Light_Detection/text.txt', 'w') as f: 
  for ann in traffic_annotations:
    image_id = ann['image_id']

    startx = ann['bbox'][0]
    starty = ann['bbox'][1]
    width = ann['bbox'][2]
    height = ann['bbox'][3]

    str_centerx = "{:.6f}".format((startx + width/2) / 640)
    str_centery = "{:.6f}".format((starty + height/2) / 480)
    str_width = "{:.6f}".format(width / w)
    str_height = "{:.6f}".format(height / h)

    if image_id in traffic_dict:
      # img_data_str = str(ann['category_id']) + " " +str(ann['bbox'][0] / w) + " " +str(ann['bbox'][1] / h) + " " +str(ann['bbox'][2] / w) + " " +str(ann['bbox'][3] / h)
      img_data_str = img_dict[image_id] + " 0 " + str_centerx + " " + str_centery + " " + str_width + " " + str_height +'\n'
      f.write(img_data_str)
      traffic_dict[image_id].append(img_data_str)
    else:
      traffic_dict[image_id] = []

print(traffic_dict)
```

```python
df = pd.DataFrame(columns = ['Image ID', 'filename', 'Start X', 'Start Y', 'Width', 'Height'])

for ann in traffic_annotations:
    image_id = ann['image_id']
    file_name = img_dict[image_id]

    startx = ann['bbox'][0]
    starty = ann['bbox'][1]
    width = ann['bbox'][2]
    height = ann['bbox'][3]

    df.loc[len(df.index)] = [image_id, img_dict[image_id], startx, starty, width, height] 

df["Image ID"] = df["Image ID"].astype(int)
```
