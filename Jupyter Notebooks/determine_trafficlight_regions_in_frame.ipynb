```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Robotics Traffic Light detections/Video/traffic_df_corrected.csv')
df
```

```python
summary = dict()

cur_frame = 0
max_x = 0
max_y = 0

max_percent_x = 0
max_percent_y = 0 

for i, row in df.iterrows():
  row = row.to_numpy()
  # print(row)
  if cur_frame == 0 or row[0] !=cur_frame:
    #save current max values if frame != 0
    if cur_frame != 0:
      summary[cur_frame]['max_x'] = max_x
      summary[cur_frame]['max_y'] = max_y

      
      summary[cur_frame]['percent_x'] = max_x * 100 // 640
      summary[cur_frame]['percent_y'] = max_y * 100 // 480

      max_percent_x = max(max_percent_x, max_x * 100 // 640)
      max_percent_y = max(max_percent_y, max_y * 100 // 480)

    cur_frame = row[0]
    summary[cur_frame] = dict()
    # summary.append(cur_frame)

    #reset max min
    max_x = 0
    max_y = 0
  
  max_x = max(max_x, row[3])
  max_y = max(max_y, row[4])

print(summary)

print("Max % x:", max_percent_x)
print("Max % y:", max_percent_y)
```
