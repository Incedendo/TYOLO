```python
import cv2

img = cv2.imread('/content/drive/MyDrive/Classroom/CSCE_452_643_Robotics_and_Spatial_Intelligence/Traffic_Light_Detection/dataset/images/000000281179.jpg')
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
# cv2.rectangle
```

```python
coordinates = ['145.42	54.89	9.84	27.83',
'163.22	35.31	8.10	20.88',
'144.69	23.72	11.10	25.95',
'79.07	20.52	10.63	18.30',
'76.08	54.82	12.55	24.86',
'108.28	22.80	7.74	30.95',
'111.26	3.86	17.62	18.46',
'144.93	46.21	8.43	11.17']

for coor in coordinates:
  vals = coor.split()

  for i in range(len(vals)):
    vals[i] = int(float(vals[i]))

  width = vals[2]
  height = vals[3]

  startx = vals[0]
  starty = vals[1]
  # 163.15	106.16	23.10	38.71
  cv2.rectangle(img, (startx, starty), (startx+width, starty+height), (255,255,255), 2)

  plt.imshow(img)
```
