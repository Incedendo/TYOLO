## Load the Json file

```python
import json

filepath = '/content/drive/MyDrive/Classroom/CSCE 452 643 Robotics and Spatial Intelligence 500,600/TrafficLight Data/labels.json'
# Opening JSON file
f = open(filepath,)
 
# returns JSON object as
# a dictionary
data = json.load(f)

w = 640
h  = 480
```

## Convert JSON into Dataframe

```python
# {
#     "id": 1,
#     "file_name": "000000000257.jpg",
#     "height": 480,
#     "width": 640,
#     "license": null,
#     "coco_url": null
# },

images = data['images']
print("no of images:", len(images))
```

```python
# {
#     "id": 88,
#     "image_id": 33,
#     "category_id": 10,
#     "bbox": [
#         143.67,
#         172.52000000000004,
#         52.349999999999994,
#         51.11
#     ],
#     "area": 2675.6085,
#     "iscrowd": 0,
#     "supercategory": "outdoor"
# },

annotations = data['annotations']
print("no of annotations:", len(annotations))

image_list = []
# Iterate over a sequence of numbers from 0 to 4
for i in range(len(images)):
    # In each iteration, add an empty list to the main list
    image_list.append([])

print("len of image_list", len(image_list))

for ann in annotations:
  image_id = ann['image_id']-1
  # print(image_id)
  img_data_str = str(ann['category_id']) + " " +str(ann['bbox'][0] / w) + " " +str(ann['bbox'][1] / h) + " " +str(ann['bbox'][2] / w) + " " +str(ann['bbox'][3] / h)
  
  image_list[image_id].append(img_data_str)

print(image_list)
```
